```python
#######################################################################################
# Part 1 - Compute and plot Julia sets


def invert_fun(complex_fun, complex_val):
    """
    Input: np.poly1d object complex_fun and complex complex_val
    
    Output: List of n possible complex solutions to the equation complex_val = complex_fun(inv)
    The roots of complex_poly are the n possible values for complex_fun^(-1)(complex_val)
    """
     
    poly_p = complex_fun - complex_val
    output = np.sort(poly_p.r)
    
    return output
```

```python
def julia_set(lmbd, z_0, num_returned, num_dropped, seed=None):
    """
    Input: complex lmbd, z_0, integers num_returned, num_dropped,
    optional integer seed
    
    Output: List of complex numbers (starting with z_0)
    generated by Julia iteration with given lmbd
    """
    output = []
    
    if seed:
        random.seed(seed)

    output.append(z_0)
    for _ in range(num_returned + num_dropped - 1):
        f_z = invert_fun(np.poly1d([lmbd * -1, lmbd, 0]), output[-1])
        output.append(random.choice(f_z))
    
    return output[num_dropped:]
```

```python
def plot_julia(julia_points, lmbd):
    """
    Input: List of complex numbers that lie on the Julia set for given lmbd
    
    Output: Scatter plot of complex number in list julia_points
    """
    
    fig = plt.subplots()
    
    real_nums = []
    imag_nums = []
    for num_i in julia_points:
        x_r = num_i.real
        y_i = num_i.imag
        real_nums.append(x_r)
        imag_nums.append(y_i)
    plt.scatter(real_nums, imag_nums)
    plt.xlabel("real axis")
    plt.ylabel("imaginary axis")
    plt.title("Julia plot for list of complex numbers that lie on the Julia set for given lmbd")
    plt.show()
        
    
    return fig
```

```python
##############################################################################
# Part 2 - Compute and plot Mandelbrot sets

MAX_ITERS = 20

def iterate_mandel(z_0):
    """
    Input: complex z_0
    
    Output: number of iterations for Mandelbrot function f(z) = z ** 2 + z_0 to diverge,
    maximum number of iterations is capped at MAX_ITERS
    """

    f_z = 0
    iters = 0
    while abs(f_z) <= 2 and iters < MAX_ITERS:
        f_z = f_z*f_z + z_0
        if not (abs(f_z) <= 2 and iters < MAX_ITERS):
            break
        iters += 1
    
    return (iters)
```

```python
def mandel_table(real_values, imag_values):
    """
    Input: arrays real_values, imag_values of floats
    
    Output: 2D numpy array indicating number of iterations for
    divergence of Mandelbrot iteration initialized at
    corresponding complex values
    """
    x_v, y_v = np.meshgrid(real_values, imag_values)

    iter_mand = np.vectorize(iterate_mandel)
    
    c_vectorized = np.vectorize(complex)

    complex_val = c_vectorized(x_v, y_v)

    output = iter_mand(complex_val)
     
    return(output)
   
```

```python
#########################################################################################
# Part 3 - Compute and plot Newton basins 

EPSILON = 0.000001

def newton_index(roots, z_0):
    """
    Input: List of complex roots, complex initial value z_0
    
    Output: Index of root in roots to which Newton iteration with initial
    value z_0 converges (with tolerance EPSILON)
    
    NOTE: If the iteration encounters a zero derivative, return len(roots)
    """
    roots_copy = copy.deepcopy(roots)
    g_z = np.poly1d(roots_copy, True)
    g_deriv_z = g_z.deriv()
    z_i = z_0
    converging = True
    while True:
        g_z_k = g_z(z_i)
        g_prime_k = g_deriv_z(z_i)
        
        if np.absolute(g_z_k) <= EPSILON: 

            break
        if np.absolute(g_prime_k) <= EPSILON:
            converging = False
            break
               
        z_i -= g_z_k/g_prime_k
        
    if not converging:
        return len(roots_copy)
    
    for num_i in range(len(roots)):
        roots_copy[num_i] = np.absolute(roots_copy[num_i] - z_i)
        
    return np.argmin(roots_copy)

            
    
            
            
```

```python
def newton_table(roots, real_values, imag_values):
    """
    Input: List of complex numbers roots, two lists of floats that define a grid
    of complex numbers
    
    Output:  2D numpy array of Newton indices where z_0 takes on values in
    the grid of complex numbers
    """
    x_v, y_v = np.meshgrid(real_values, imag_values)

    # iter_newt = np.vectorize(newton_index)
    
    c_vectorized = np.vectorize(complex)
    
    complex_val = c_vectorized(x_v, y_v)
    
    imag_len = len(complex_val)
    real_len = len(complex_val[0])
    arr_dim = (imag_len, real_len)

    output = np.zeros(arr_dim)
    
    for num_i in range(imag_len):
        for num_j in range(real_len):
            output[num_i][num_j] = newton_index(roots, complex_val[num_i][num_j])
            
    return(output).astype(int)
```

```python
#########################################################################################
# Student code for plotting Newton basins (peer-graded)

def plot_newton(roots, real_values, imag_values):
    """
    Input: List of complex numbers roots, two lists of floats that define a grid
    of complex numbers
    
    Output: matplotlib figure of mage plot of mandel_table()
    applied to roots, real_values, imag_values
    """
    
    fig = plt.subplots()

    x_real = real_values
    y_imag = imag_values
    roots_c = copy.deepcopy(roots)
    z_n = newton_table(roots_c, x_real, y_imag)

    real_nums = []
    imag_nums = []
    for num_i in roots:
        x_r = num_i.real
        y_i = num_i.imag
        real_nums.append(x_r)
        imag_nums.append(y_i)
    plt.scatter(real_nums, imag_nums, c='red')
    plt.xlabel("real axis")
    plt.ylabel("imaginary axis")
    plt.title("Newton basin for real and imaginary values with " + str(len(roots)) + " roots")
    
    plt.imshow(z_n)
    plt.show()
    
    return fig
```
